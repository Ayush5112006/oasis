# 1.Data Loading And Preprocessing

```python
# First We Need To Import Libraries For Further Preprocessing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Now We See The Data

menu_dataset = pd.read_csv('menu.csv')
```

```python
# now We Clean the dataset

if 'Unnamed: 0' in menu_dataset.columns:
    menu_dataset.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# Check Whether any null or missing values is there

menu_dataset.isnull().sum()
```

```python
# check Whether any redundant rows

menu_dataset.duplicated().sum()
```

```python
# check whether the datatypes is correct

menu_dataset.info()
```

```python
# now here we see the serving size that have string so we convert into float

# also removing strings
import re

def extract_serving_size(serving_size_str):
    match = re.search(r'(\d+\.?\d*)', serving_size_str)
    if match:
        return float(match.group(1))
    return None

menu_dataset['Serving Size'] = menu_dataset['Serving Size'].apply(extract_serving_size)
```

# 5. Visualization: Present insights through bar charts, line plots, and heatmaps.

```python
# bar charts

menu_dataset['Category'].value_counts().plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Category Distribution')
plt.show()
```

```python
# Heatmap

# Define columns to include for a less redundant correlation heatmap
# Excluding 'Calories' and 'Calories from Fat' as requested, and prioritizing main nutrient values
selected_cols = [
    'Serving Size',
    'Total Fat', 'Saturated Fat', 'Trans Fat',
    'Cholesterol',
    'Sodium',
    'Carbohydrates', 'Dietary Fiber', 'Sugars',
    'Protein',
    'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)'
]

# Filter the dataset to include only the selected numeric columns
filtered_numeric_menu_dataset = menu_dataset[selected_cols]

correlation_matrix = filtered_numeric_menu_dataset.corr()

plt.figure(figsize=(12, 10)) # Increased figure size for better readability with more labels
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('Filtered Correlation Heatmap of Nutrient Values')
plt.show()
```

# 6. Recommendations

recommendations based on the EDA:

-->Nutritional Balance: Several items, particularly from 'Desserts' and 'Smoothies & Shakes' categories, show very high calorie, fat, and sugar content. Consider offering healthier/lighter versions or smaller portion sizes to cater to health-conscious customers.

-->Menu Diversification: Categories like 'Salads' and 'Desserts' have a significantly lower number of items. Expanding the variety in these categories could attract a broader customer base or encourage existing customers to explore more options, potentially increasing sales in these segments.

-->Ingredient Analysis: The heatmap showed strong correlations between various fat components and between carbohydrates and sugars. This suggests that adjustments in core ingredients (e.g., opting for leaner proteins, reducing added sugars) could significantly impact overall nutritional profiles across multiple menu items simultaneously.

-->Promotional Focus: Given that 'Coffee & Tea' and 'Breakfast' are the most popular categories, focus on maintaining their appeal through loyalty programs or seasonal specials. For less popular categories like 'Salads', targeted promotions could help increase their visibility and demand.
